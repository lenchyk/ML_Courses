### Your task:

- Read data from **train.csv**  
Form new dataset from rows (houses) with high OverallQual (8,9,10) and following columns:  
'LotArea', 'YearBuilt', 'LotShape', 'HouseStyle', 'OverallCond', 'Heating', 'KitchenQual',   
'Functional', 'YearRemodAdd', 'GarageCars', 'GarageArea', 'GarageQual','GarageFinish','PoolArea',  
 'SalePrice'

- 
1) Find average price and count of house sold, grouping by YearBuilt (create dataset with these results)      
2) Create new column which reflects number of years since remodelling  
3) Transform column LotArea: square feet -> square meters  
4) Create new column which group LotArea into 3 bins: Small, Medium, Large (set thresholds of your choise)  
5) Are missing values in data? If so, fill them by some values   

- Provide in-depth analysis of the dataset (find as many insights in data as possible ;))

```python
import pandas as pd
import numpy as np
import random
```

```python
data = pd.read_csv("train.csv")
```

```python
raw_data = data[data.OverallQual > 8]
data1 = pd.DataFrame(raw_data, columns=['OverallQual', 'LotArea', 'YearBuilt', 'LotShape', 'HouseStyle', 'OverallCond', 'Heating', 'KitchenQual',
'Functional', 'YearRemodAdd', 'GarageCars', 'GarageArea', 'GarageQual','GarageFinish','PoolArea',
'SalePrice'])
data1
```

```python
# (1) average price and amount of houses sold grouped by YearBuilt
avg = data1.groupby("YearBuilt").mean()
sold = data1.count()
avg
```

```python
sold
```

```python
# (2) creating new column with years since remodeling
data1["AfterRemod"] = 2019 - data1["YearRemodAdd"]
data1
```

```python
# (3) square feet --> square meters
data1["LotArea"] = (0.3048 * 0.3048) * data1.LotArea
data1
```

```python
# (4) grouping LotArea into 3 bins: Small (0-100), Medium (101-250) and Large(all others)
data1["LotAreaMore"] = data1.apply(lambda row: "Small" if ((row.LotArea>0) & (row.LotArea <= 100))
                                                else ("Medium" if ((row.LotArea>100) & (row.LotArea <= 250))
                                                               else "Large"), axis=1) 
```

```python
# (5) filling NaN with some numbers
data1.fillna(100, inplace=True)
```

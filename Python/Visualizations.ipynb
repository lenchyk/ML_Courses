### Exercises

1) Draw graphs of y = sin(x) and y=2*cosx -1

```python
x = np.linspace(-2 * math.pi, 2 * math.pi, 100)
y1 = np.sin(x)
y2 = 2 * np.cos(x) - 1
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4), sharey=True, dpi=200)
ax1.plot(x, y1, 'g-')
ax2.plot(x, y2, 'b-')
ax1.set_title("y = sin(x)"); ax2.set_title("y = 2 * cos(x) - 1")
ax1.set_xlim(-2 * math.pi, 2 * math.pi)
ax2.set_xlim(-2 * math.pi, 2 * math.pi)
ax1.set_ylim(-1, 1)
ax2.set_ylim(-3, 2)

ax1.grid(linestyle='--'); ax2.grid(linestyle='--')
plt.tight_layout()
plt.show()
```

2) Draw scatter plot to visualize dependency between YearBuilt, LotArea, EnclosedPorch

```python
figure = plt.figure(figsize=(10, 6), dpi=80, facecolor='w', edgecolor='k')
plt.scatter("LotArea", "EnclosedPorch", data=data, c="YearBuilt", cmap="Reds", 
            edgecolors='black', linewidth=.5)
plt.title("LotArea and YearBuilt vs EnclosedPorch", fontsize=16)
plt.xlabel("LotArea", fontsize=10)
plt.ylabel("EnclosedPorch", fontsize=10)
cbar = plt.colorbar()
cbar.set_label("YearBuilt", rotation=270)
plt.show()
```

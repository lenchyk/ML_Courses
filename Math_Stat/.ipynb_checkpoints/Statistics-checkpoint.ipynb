### Visualizations

```python
sns.set()
```

```python
# bar chart
sns.countplot(x='HouseStyle', data=data)
plt.title('HouseStile Bar Chart')
plt.show()
```

```python
# histogram
figure = plt.figure(figsize=(8,6))
ax = data.SalePrice.hist(bins=20, alpha=0.85, rwidth=0.95)
plt.xlabel('SalePrice')
plt.ylabel('Frequency')
plt.title('SalePrice histogram', fontsize=14)
plt.show()
```

```python
# boxplot
figure = plt.figure(figsize=(4,6))
ax = sns.boxplot(data.SalePrice, orient='vertical', color='g')
plt.title('SalePrice Boxplot')
plt.show()
```

```python
figure = plt.figure(figsize=(8,6))
ax = sns.boxplot(x='OverallQual', y="SalePrice", data=data[['OverallQual', 'SalePrice']])
plt.title('SalePrice Boxplots by OverallQual')
plt.show()
```

```python
figure = plt.figure(figsize=(8,6))
ax = plt.scatter(data.SalePrice, data.LotArea, c="green", alpha=0.5, marker='o')
plt.xlabel("SalePrice")
plt.ylabel("LotArea")
plt.title('Scatter plot')
plt.show()
```

```python
# scatter plot
figure = plt.figure(figsize=(8,6))
ax = plt.scatter(data.SalePrice, data.OverallCond, marker='*')
plt.xlabel("SalePrice")
plt.ylabel("OverallCond")
plt.title('Scatter plot')
plt.show()
```

```python
figure = plt.figure(figsize=(8,6))
ax = sns.scatterplot('SalePrice', 'LotArea', hue='OverallQual', 
                      data=data, palette='OrRd',  legend='full')
plt.xlabel("SalePrice")
plt.ylabel("LotArea")
plt.title("Scatter plot", fontsize=14)
plt.show()
```

```python
# correlations heatmap
corrmat = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corrmat, square=True)
plt.show()
```

```python
cols1 = corrmat.nlargest(10, 'SalePrice').index.values
corrmat_largest = data[cols1].corr()
figure = plt.figure(figsize=(8, 7))
sns.heatmap(corrmat_largest, square=True, annot=True)
plt.show()
```

```python
# pairplot
cols = ['YearBuilt', 'OverallCond', 'GrLivArea']
sns.pairplot(data[cols])
plt.show()
```
